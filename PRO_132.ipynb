```python
import pandas as pd
df=pd.read_csv('/content/PRO-131/star_with_gravity.csv')
```

```python
df
```

```python
mass=df["Mass"].to_list()
radius=df["Radius"].to_list()
gravity=df["Gravity"].to_list()
```

```python
mass.sort()
radius.sort()
gravity.sort()
import matplotlib.pyplot as plt
plt.plot(radius,mass)
plt.title("radius and mass of the star")
plt.xlabel("Radius")
plt.ylabel("Mass")
plt.show()
```

```python
plt.plot(gravity,mass)
plt.title("gravity and mass of the star")
plt.ylabel("Gravity")
plt.xlabel("Mass")
plt.show()
```

```python
plt.plot(gravity,radius)
plt.title("gravity and radius of the star")
plt.ylabel("Gravity")
plt.xlabel("Radius")
plt.show()
```
